```python
import google.generativeai as genai

# Set your Gemini API key
genai.configure(api_key="GEMINI_API_KEY")

# Load Gemini model
model = genai.GenerativeModel('gemini-2.0-flash')

# Start a chat session
chat = model.start_chat(history=[])
```

```python
import json
import pandas as pd

chat_log = []

def log(message, sender="agent"):
    chat_log.append({"sender": sender, "message": message})
    print(f"{sender.capitalize()}: {message}")

def get_user_input(prompt):
    log(prompt)
    user_input = input("You: ")
    log(user_input, sender="user")
    return user_input

def get_gemini_reply(user_message):
    response = chat.send_message(user_message)
    reply = response.text
    log(reply)
    return reply

def main():
    log("Hello! I'm your AI Sales Agent. Let's find the perfect product for you.")

    name = get_user_input("Can I know your name?")

    need = get_user_input(f"Nice to meet you, {name.title()}! Are you looking for a **phone** or a **SaaS plan**?")
    get_gemini_reply(f"The user is looking for {need}. Recommend a product with pricing and features.")

    budget = get_user_input("What’s your budget in USD?")
    get_gemini_reply(f"The user has a budget of {budget} dollars. Suggest suitable options for {need}.")

    feature = get_user_input("Any specific feature you care about? (e.g., camera, storage, email)")
    get_gemini_reply(f"The user wants features like {feature}. Suggest best products matching these preferences.")

    objection = get_user_input("Do you have any concerns or questions?")
    get_gemini_reply(f"The user has '{objection}' concern. Address it politely and convincingly.")

    log("Thanks for the chat! We’ll follow up soon with more info. Logging this interaction for our team.")
    save_log(name)

def save_log(customer_name):
    df = pd.DataFrame(chat_log)
    df.to_csv(f"{customer_name}_gemini_chat.csv", index=False)
    with open(f"{customer_name}_gemini_chat.json", "w") as f:
        json.dump(chat_log, f, indent=2)

if __name__ == "__main__":
    main()
```
